```python
def show_png(out_dir, short_window, long_window):
    for dirpath, dirnames, filenames in os.walk(out_dir):
        for filename in filenames:
            title = "MACDStrategy" + "_" + str(short_window) + "_" + str(long_window)
            if filename[0:len(title)] == title:
                img = mpimg.imread(os.path.join(dirpath, filename))
                plt.figure(figsize = (52, 26))
                plt.imshow(img)
                plt.axis('off')
                plt.show()
#                 img = Image.open(os.path.join(dirpath, filename))                
#                 img.show()
#                 img.format
                return 
```
